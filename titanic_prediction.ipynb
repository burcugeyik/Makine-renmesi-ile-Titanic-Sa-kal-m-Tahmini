```python
import pandas as pd
import numpy as np
import xgboost as xgb

# 1. Kaggle'dan indirdiğimiz eğitim ve test dosyalarını okuyoruz
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")

# 2. İki veri setini de aynı anda temizlemek için bir döngü kuruyoruz
for df in [train_df, test_df]:
    df['Age'] = df['Age'].fillna(train_df['Age'].mean()) # Train ortalamasıyla doldur
    df['Fare'] = df['Fare'].fillna(train_df['Fare'].mean())
    df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
    df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# 3. Model Girdilerini Seçme
ozellikler = ['Pclass', 'Sex', 'Age', 'Fare', 'FamilySize']

X_train = train_df[ozellikler]
y_train = train_df['Survived'] # Train'de sonuçlar var
X_test = test_df[ozellikler]   # Test'te sonuç yok, biz tahmin edeceğiz

# 4. Modeli Eğitme (Tüm veriyle eğitiyoruz ki daha güçlü olsun)
model = xgb.XGBClassifier(random_state=42)
model.fit(X_train, y_train)

# 5. Kaggle'ın bizden beklediği test verisi tahminlerini yapıyoruz
tahminler = model.predict(X_test)

# 6. KAGGLA'A YÜKLENECEK DOSYAYI OLUŞTURMA (En Kritik Adım)
# Kaggle bizden sadece PassengerId ve Survived kolonlarını ister
submission = pd.DataFrame({
    "PassengerId": test_df["PassengerId"],
    "Survived": tahminler
})

# Dosyayı csv olarak kaydediyoruz
submission.to_csv("Benim_Ilk_Tahminim.csv", index=False)
print("🏆 Harika! 'Benim_Ilk_Tahminim.csv' dosyası başarıyla oluşturuldu. Kaggle'a yüklemeye hazır!")
```
